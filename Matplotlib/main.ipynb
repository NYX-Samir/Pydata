### BAR PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["Python","C++","Java","JavaScript"]
y=[85,60,78,89]
z=[20,10,30,40]
plt.xlabel("Langauage",fontsize=15)
plt.ylabel("Popularity",fontsize=15)
plt.title(2025,fontsize=30)
#colors=['r','b','y','g']
plt.bar(x,y,width=0.5,color='r',edgecolor='black',label="Popularity")
plt.bar(x,z,width=0.5,color='black',edgecolor='black',label="Users")
plt.legend()
plt.show()
```

```python
categories = ["A", "B", "C", "D", "E"]  
values1 = [10, 20, 15, 25, 30]  # First dataset  
values2 = [15, 10, 25, 20, 35]  # Second dataset  

bar_width=0.3

X=np.arange(len(categories))

plt.title("Two Dataset")
plt.bar(X-bar_width/2,values1,color="b",width=bar_width,label="DATASET 1")
plt.bar(X+bar_width/2,values2,color="r",width=bar_width,label="DATASET 2")
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,6))
ax1=fig.add_subplot(3,3,1)
ax2=fig.add_subplot(3,3,3)
ax3=fig.add_subplot(3,3,5)
ax1.hist(np.random.randn(100),bins=20,color='b')
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30),marker='*')
ax3.bar(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
tick=ax.set_xticks([0,200,400,600,800,1000])
labels=ax.set_xticklabels(["One","Two","Three","Four","Five","Six"],rotation=30,fontsize="small",color="b")
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),"k-",label="one")
ax.plot(np.random.randn(1000).cumsum(),"k.-",label="Two")
ax.plot(np.random.randn(1000).cumsum(),"k.",label="Three")
plt.legend(loc="best")
```
