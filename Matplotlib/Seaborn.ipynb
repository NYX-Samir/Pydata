```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\evilk\Downloads\archive\supermarket_sales - Sheet1.csv')
print(df.head(10))
```

```python
sns.displot(df["gross income"],kde=True,bins=40)
```

### Joint 

```python
sns.jointplot(x="Tax 5%",y='Quantity',data=df,kind="hex")
sns.jointplot(x="Tax 5%",y='Quantity',data=df,kind="kde")
sns.jointplot(x="Tax 5%",y='Quantity',data=df,kind="reg")
```

### KDE


```python
sns.kdeplot(df["Rating"])
```
